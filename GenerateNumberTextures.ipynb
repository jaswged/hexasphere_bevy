```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np
```

```python
message = "Hello, World!"
```

```python
image = Image.new("RGB", (256, 256), "blue")
```

```python
draw = ImageDraw.Draw(image)
```

```python
font = ImageFont.truetype("arial.ttf", size=40)
```

```python
draw.text((10, 10), message, font=font, fill=(255, 255, 255))
```

```python
image.save("result.png")
```

```python
def save_num_img(num):
    color = tuple(np.random.choice(range(256), size=3))
#     print(color)
    image = Image.new("RGB", (256, 256), color)  #"blue")
    draw = ImageDraw.Draw(image)
    font = ImageFont.truetype("arial.ttf", size=145)
    
    text_width, text_height = draw.textsize(str(num), font=font)
    x_text = (256 - text_width) // 2
    y_text = (256 - text_height) // 2
    
#     draw.text((10, 10), str(num), font=font, fill=(255, 255, 255))
    draw.text((x_text, y_text), str(num), font=font, fill=(255, 255, 255))
    image.save(f"assets/num_textures/{num}.png")
```

```python
save_num_img(5)
```

```python
# Unity has 92 tiles
# Unity_4 has 162 tiles
# Unity_6 has 362 tiles
# Unity_9 has 812 tiles
# Unity_10 has 1002 tiles
# Unity_20 has 4002 tiles
for i in range(0, 1000):
    save_num_img(i)
```
